```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
```

```python
url = 'https://en.wikipedia.org/wiki/2022_FIFA_World_Cup'

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')
soup
```

```python
events = soup.find_all('table', class_='fevent')

home = []
score = []
away = []

for event in events:
    home.append(event.find('th', class_='fhome').get_text())
    score.append(event.find('th', class_='fscore').get_text())
    away.append(event.find('th', class_='faway').get_text())

df_football = pd.DataFrame({'home': home, 'score': score, 'away': away})
df_football
```
